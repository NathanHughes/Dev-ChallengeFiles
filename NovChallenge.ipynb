```python
#helpers

def call_query(name):
    try:
        
        wiki =wikipedia.page(name)
        url = wiki.url
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')
        wiki_Cheat(soup)
    except wikipedia.exceptions.DisambiguationError as e:
        call_query(e.args[1][0])
    except wikipedia.exceptions.RedirectError as e:
        # (pageid=None, *args)
        print (e)
        
    except wikipedia.exceptions.PageError as e:
        # (pageid=None, *args)
        print (e)


def check_cat(check):
    for cat, result in wikiCats.items():    
        if check.lower().find(cat)>=0:
            return result

def parse_biota(soup, biota):
    rows = biota.findAll('tr')
    infobox = {}

    for row in rows:
        try:
            tds = row.select('td')
            infobox[tds[0].text.strip()] = tds[1].next_element.text.strip()

        except:
            continue

    if 'Kingdom:' in infobox:
        outputCat = kingdoms[infobox['Kingdom:']] 
        print( inputName + ' is ' + types[outputCat] )
    else:
        get_name(soup)
        
def get_name(page):
    vernacular = page.find(class_='vernacular')
    binominal  = page.find(class_='binominal')

    if vernacular is not None:
        a = vernacular.findAll('a', href=True)
        linkParts = a[0]['href'].split('/')
        call_query(linkParts[-1])
    elif binominal is not None:
        a = binominal.findAll('a', href=True)
        linkParts = a[0]['href'].split('/')
        call_query(linkParts[-1])
        
    else:
        classification = page.find(id='mw-normal-catlinks')
        cats = classification.findAll('li')
        findCats(cats)

        
def findCats(cats):

    foundCat = False
    for cat in cats:
        result =check_cat(cat.text)
        if result is not None:
            print(inputName + " is " + types[result])
            foundCat = True
            break

    if foundCat is False:
        print("Unknown")
```

```python
def wiki_Cheat(soup):
    links =[]
    biota = soup.find("table",{"class":"biota"})
    classification = soup.find(id='mw-normal-catlinks')
    #print(classification)
    if biota is not None:
        parse_biota(soup, biota)

    elif biota is None and classification is not None:

        cats = classification.findAll('li')

        if cats[0].text.lower().find('disambiguation page')>=0:
            
            links = soup.findAll('li',attrs={'class': None})
            firstLink = links[0].find('a')['href'].split('/')

            call_query(firstLink[-1])
        else:
            findCats(cats)
```
